```python
import numpy as np
import pandas as pd
```

```python

file = pd.read_csv('titanic.csv')
```

```python
file
```

```python
file.dtypes
```

```python
# Remove columns which have no impact on the survival of a person.
del file['Name']
del file['Ticket']
del file['Fare']
```

```python
file
```

```python
file.isnull().sum()
```

```python
# it is observed that out of 891 , column cabin have 687 null so we will drop cabin column as well.
del file['Cabin']
file.head()
```

```python
# Embarked is column where we need to apply most frequncy to deal the 2 null ones.

file['Embarked'].unique()
```

```python
file['Embarked'].value_counts()
```

```python
file['Embarked'].replace(np.nan,'S',inplace=True)
```

```python
# dealing null values in age 
age_mean = file['Age'].mean()
file['Age'].replace(np.nan,age_mean,inplace=True)
```

```python
file.info()
```

```python
# dealing categorical value of sex 
def getNumber(str):
    if str=="male":
        return 1
    else:
        return 2

file['Sex'] = file["Sex"].apply(getNumber)
```

```python
file.head()
```

```python
# dealing categorical value of Embarked
# one hot-encoding
Embarked_dummy=pd.get_dummies(file['Embarked'],dtype=int)
Embarked_dummy
```

```python
file=pd.concat([file,Embarked_dummy],axis=1)
file.head()
```

```python
# delete Embarked with categorical data.
del file['Embarked']
file.head()
```

```python
# find feautures and output.
X = file.iloc[:,2:11]
Y = file.iloc[:,1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state = 0)
```

```python
from sklearn import tree
from sklearn.metrics import confusion_matrix
clf = tree.DecisionTreeClassifier()
clf.fit(x_train, y_train)
Y_pred_l = clf.predict(x_test)
clf.score(x_train,y_train), clf.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth = 6, random_state=0)
clf.fit(x_train, y_train)
y_test_pred = clf.predict(x_test)
clf.score(x_train,y_train),clf.score(x_test,y_test)
```

```python
confusion_matrix(y_test,y_test_pred)
```
